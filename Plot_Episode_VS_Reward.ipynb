```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
dfPPO = pd.read_csv('PPO/PPO_record.csv')
dfSAC = pd.read_csv('SAC/SAC_reocrd_f.csv')
dfTD3 = pd.read_csv('TD3/TD3_record.csv')
```

```python
from scipy.ndimage.filters import gaussian_filter1d
dfPPO = dfPPO.apply(lambda x: gaussian_filter1d(x, sigma=3))
dfSAC = dfSAC.apply(lambda x: gaussian_filter1d(x, sigma=3))
dfTD3 = dfTD3.apply(lambda x: gaussian_filter1d(x, sigma=3))
```

```python
sns.lineplot(data=dfPPO, x='episode', y='reward')
plt.savefig('e_vs_rPPO.png', dpi=300)
```

```python
sns.lineplot(data=dfSAC, x='episode', y='reward')
plt.savefig('e_vs_rSAC.png', dpi=300)
```

```python
sns.lineplot(data=dfTD3, x='episode', y='reward')
plt.savefig('e_vs_rTD3.png', dpi=300)
```

```python
dfPPO = dfPPO.iloc[:63, :]
dfSAC = dfSAC.iloc[:750, :]
dfTD3 = dfTD3.iloc[:750, :]
```

```python
plt.plot(dfPPO['episode'], dfPPO['reward'], label = "PPO")
plt.plot(dfSAC['episode'], dfSAC['reward'], label = "SAC")
plt.plot(dfTD3['episode'], dfTD3['reward'], label = "TD3")
plt.legend()
plt.show()
```
