```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
df = pd.read_csv('PPO_record.csv')
```

```python
from scipy.ndimage.filters import gaussian_filter1d
df = df.apply(lambda x: gaussian_filter1d(x, sigma=3))
```

```python
sns.lineplot(data=df, x='episode', y='reward')
plt.savefig('e_vs_r.png', dpi=300)
```
